```python
%matplotlib inline
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np
import os
from IPython.display import display
```

```python
filename = 'Sessões com os membros 2022_1 (respostas).xlsx'
if filename in os.listdir():
    print(f'Found the file {filename}, ready to go!')
else:
    print(f'No file found with {filename} in directory {os.getcwd()}, have you downloaded the file?')
```

```python
dataframe = pd.read_excel(filename)
dataframe
```

```python
questions = list(dataframe.head(0))

for _, row in dataframe.iterrows():
    row = row.values.tolist(); nome = row[1]
    nome.replace(' ','_')
    print(len(row))
    with open(f'devolutivas/{nome}.txt', 'w') as file:
        for i in range(1,len(questions)):
            file.write(f'{questions[i]}: {row[i]}\n\n')
```
